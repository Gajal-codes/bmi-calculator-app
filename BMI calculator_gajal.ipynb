```python
def calculate_bmi():
    print("🧮 Welcome to the BMI Calculator App!")

    # Step 1: Name input
    name = input("👤 What's your name? ")

    # Step 2: Weight input
    weight_input = input("⚖️ Enter your weight in kilograms (e.g., 48): ")

    # Step 3: Height input
    height_input = input("📏 Enter your height in centimeters (e.g., 152): ")

    try:
        weight = float(weight_input)
        height_cm = float(height_input)
        height_m = height_cm / 100

        # Step 4: BMI calculation
        bmi = weight / (height_m ** 2)
        bmi = round(bmi, 2)

        # Step 5: Output
        print(f"\n👋 Hello {name}, your BMI is: {bmi}")

        if bmi < 18.5:
            print("📉 You are underweight.")
        elif 18.5 <= bmi < 25:
            print("✅ You have a normal weight.")
        elif 25 <= bmi < 30:
            print("⚠️ You are overweight.")
        else:
            print("🚨 You are obese. Please take care!")

    except ValueError:
        print("❌ Please enter valid numbers for weight and height!")

# Run the function
calculate_bmi()
```
